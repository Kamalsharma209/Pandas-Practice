```python
import pandas as pd
var = pd.DataFrame({"a":[23,45,90,76,65],"b":[56,90,34,12,50],"c":[99,101,3,109,89]})
var
```

**Insert data between two columns **

```python
var.insert(2, "d", [99, 90, 23, 12, 1])
var
```

**insert by slicing**

```python
var1 = pd.DataFrame()
var1["e"] = var["d"][:3]
var1
```

**Delete an dataframe**

```python
var.pop("c")
```

```python
var
```

**By delete method**

```python
del var["b"]
var
```

**Create Sample DataSet**

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'F', 'M'],
    'Subject': ['Math', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science', 'English'],
    'Score': [85, 78, 92, 66, 88, 75, 95, 90, 87, 82],
    'Section': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
}

df = pd.DataFrame(data)
df
```

**Group by One Columns**

```python
df.groupby('Gender')['Score'].mean()
```

**Group by Multiple Columns**

```python
df.groupby(['Gender','Subject'])['Score'].mean()
```

**Use Aggregation Function**

```python
print(df.groupby('Subject')['Score'].sum())
print('\n Number of Students per Subject')
print(df.groupby('Subject')['Name'].count())
```

**Using .agg() for Multiple Aggregation**

```python
df.groupby ('Subject')['Score'].agg(['mean','min','max'])
```
